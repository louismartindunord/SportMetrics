```python
from dotenv import load_dotenv

def create_connection():
    load_dotenv()

    DATABASE = os.getenv("DATABASE")
    HOST = os.getenv("HOST")
    USER = os.getenv("USER")
    PASSWORD = os.getenv("PASSWORD")
    PORT = os.getenv("PORT")
    connection = psycopg2.connect(
        host=HOST, database=DATABASE, user=USER, password=PASSWORD, port=PORT
    )
    return connection
```

## Extraction et mise en forme des données 

```python
df = pd.read_csv('../datas/data_sport.csv', delimiter=',',on_bad_lines='skip', decimal='.')
df.head()
```

```python
def clean_data(x): 
    x = str(x)
    x = x.replace(',','.')
    x = x.replace('\xa0', '')
    return float(x)

```

```python

df["poid"] = df["poid"].apply(clean_data) 
df["poid"] = df["poid"].astype("float")
df["Date"] =  pd.to_datetime(df["Date"])

```

```python
df["Nbr_de_repetitions"] = df["Nbr_de_repetitions"].replace('max',5 )
df["Nbr_de_repetitions"] = df["Nbr_de_repetitions"].apply(clean_data) 
df["Nbr_de_repetitions"] = df["Nbr_de_repetitions"].astype("float")
```

```python
# correction of ValueError: could not convert string to float: '1h30'
df["Durée"] =  df["Durée"].replace('1h30', 90)
```

```python
df["Durée (min)"] = df["Durée"].replace("Nan", np.nan)
df["Durée (min)"] = df["Durée"].replace("7,2", 7.2)
df["Durée"] = df["Durée"].astype("float")
```

```python
df = df[["Date","Sport","Séance","Exercice","Durée","poid","Nbr_de_repetitions",'Commentaire']]
```

```python
df["id"] = df.index
```

```python
# Insérer les données du DataFrame dans la base de données
def insert_data_to_db(df):
    connection = create_connection()
    cursor = connection.cursor()
    insert_query = """
            INSERT INTO musculation_exercices (musculation_exercice_name, muscle_area)
            VALUES (%s, %s)
        """
    
    for index, row in df.iterrows():
        
        cursor.execute(insert_query, (row['Exercice'], row['Muscle_Area']))
    
    connection.commit()
    cursor.close()
    connection.close()

# Appeler la fonction pour insérer les données
insert_data_to_db(df_musculation_area)
```
